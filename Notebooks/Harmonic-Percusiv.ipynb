Percussive-to-harmonic ratio (PHR) is a measure that describes the relative balance between the percussive and harmonic components of a sound. It is computed by dividing the energy of the percussive component by the energy of the harmonic component.

es una medida adimensional, ya que representa una relación entre dos medidas de energía. Por lo tanto, no tiene unidades físicas específicas. El PHR se expresa típicamente como un número decimal entre 0 y 1, donde 0 representa una señal totalmente armónica y 1 representa una señal totalmente percusiva.


```python
def PHR(H_spec, P_spec):
    """
    Calcula el percussive-to-harmonic ratio adimensional
    """
    # Calcula el contrastre espectral para cada frecuencia
    # El componente armónico tiene mayor energía en las frecuencias fundamentales.
    # El componente percusivo tiene mayor energía en los componentes transitorios.
    contrast_H = librosa.feature.spectral_contrast(S=np.abs(H_spec))
    contrast_P = librosa.feature.spectral_contrast(S=np.abs(P_spec))
    # Compute the PHR as the ratio of percussive to harmonic energy
    PHR = contrast_P.mean() / contrast_H.mean()
    return PHR
```
